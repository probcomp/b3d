```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
clicks = []

def onclick(event):
    # Collect the click coordinates
    print(event)
    if event.xdata is not None and event.ydata is not None:
        clicks.append((event.xdata, event.ydata))
        print(f'You clicked at: ({event.xdata}, {event.ydata})')
```

```python
%matplotlib ipympl
# https://matplotlib.org/ipympl/examples/full-example.html
import numpy as np

# Load an image
# img = mpimg.imread('your_image.png')  # Replace 'your_image.png' with your image file
img = np.random.rand(100, 100)

# Display the image
fig, ax = plt.subplots()
ax.imshow(img)

# Connect the click event
cid = fig.canvas.mpl_connect('button_press_event', onclick)

# Show the image
plt.imshow(img)
```
