```python
import json
from os import listdir
from os.path import isfile, join
```

```python
first_all = 'gt'
# path = f'/home/haoliangwang/data/b3d_tracking_results/all_scale/{first_all}'
path = f'/home/haoliangwang/data/b3d_tracking_results/test'
scenario = 'collide'
scenario_path = f'{path}/{scenario}'
onlyjson = [
    f
    for f in listdir(scenario_path)
    if isfile(join(scenario_path, f)) and join(scenario_path, f).endswith(".json")
]
```

```python
configs = set([f[:-10] for f in onlyjson])
```

```python
for config in configs:
    config_dict = {}
    for f in onlyjson:
        if f.startswith(config):
            print(f)
            index = int(f[-9:-5])
            f_path = join(scenario_path, f)
            with open(f_path) as f:
                this_trial = json.load(f)
                config_dict[index] = this_trial
    with open(f"{path}/{config}.json", "w") as f:
        json.dump(config_dict, f)
```
